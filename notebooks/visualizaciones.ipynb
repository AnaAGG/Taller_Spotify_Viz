| Característica | Significado |   
|---|---|
|**Danceability**   |  Describe lo adecuada que es una pista para bailar basándose en una combinación de elementos musicales como el tempo, la estabilidad del ritmo, la fuerza del compás y la regularidad general. Un valor de 0 es el menos bailable y 1 el más bailable. 
| **Acousticness**  |  Una medida de 0 a 1 que indica si la pista es acústica. |
|**Energy**   |  Es una medida de 0 a 1 y representa una medida perceptiva de intensidad y actividad. Normalmente, las pistas energéticas son rápidas, fuertes y ruidosas.| 
|**Instrumentalness** |Predice si una pista no contiene voces. Cuanto más se acerque el valor de instrumental a 1,0, mayor será la probabilidad de que la pista no contenga voces.   |  
| **Liveness**| Detecta la presencia de público en la grabación. Los valores más altos representan una mayor probabilidad de que la pista se haya grabado en directo.   |   
| **Loudness**| La sonoridad general de una pista en decibelios (dB). Los valores de sonoridad se promedian en toda la pista. Los valores típicos oscilan entre -60 y 0 dB.  |  
|**Speechiness** | La sonoridad detecta la presencia de palabras habladas en una pista. Cuanto más exclusivamente hablada sea la grabación (por ejemplo, un programa de entrevistas, un audiolibro o poesía), más se acercará a 1 el valor del atributo.  |  
| **Tempo** |  En terminología musical, el tempo es la velocidad o el ritmo de una pieza determinada y se deriva directamente de la duración media de los tiempos.   | 
| **Valence**| De 0 a 1 que describe la positividad musical que transmite una pista. Las pistas con valencia alta suenan más positivas (por ejemplo, felices, alegres, eufóricas), mientras que las pistas con valencia baja suenan más negativas (por ejemplo, tristes, deprimidas, enfadadas).    |

## ¿Qué década es la que mejor música nos ha dado? 

Para responder a esta pregunta necesitaremos:

- Sacar de la columna de `date` los años en los que fueron lanzadas las canciones. 

- Calcular la década a la que pertencen cada una de las canciones. 

- Calcular cuántas canciones hay por década.

```python
df_sin_duplicados.head()
```

```python
# convertimos la columna de 'date' a tipo datetime
df_sin_duplicados["date"] = pd.to_datetime(df_sin_duplicados["date"])

# creamos una columna nueva donde solo tengamos el año
df_sin_duplicados["year"] = df_sin_duplicados["date"].dt.year

# calculamos la década a la que pertenece cada cancion
df_sin_duplicados["year"] = df_sin_duplicados['year'] - df_sin_duplicados['year'] % 10
```

```python
df_sin_duplicados.head()
```

```python
df_mejor_decada_pop = df_sin_duplicados.groupby("year")["popularity"].mean().reset_index()
df_mejor_decada_pop
```

```python
plt.figure(figsize = (20, 5))
sns.barplot(x = "year", y = "popularity", data = df_mejor_decada_pop, palette= "viridis", edgecolor = "black");
```

```python
df_artistas = pd.read_csv("../data/artistas_taller.csv" )
df_artistas.head()
```

```python
df_decada.head(1)
```

```python
artistas_decada = df_decada["artist"].unique().tolist()
artistas_decada[:5]
```

```python
df_artistas_decada = df_artistas[df_artistas["artist_name"].isin(artistas_decada)]
df_artistas_decada.head()
```

```python
import ast
df_artistas_decada["genres"] = df_artistas_decada["genres"].apply(ast.literal_eval)
```

```python
df_artistas_decada = df_artistas_decada.explode("genres")
df_artistas_decada.head()
```

```python
df_artistas_decada.isnull().sum()
```

```python
df_artistas_decada.dropna(inplace = True)
```

```python
from wordcloud import WordCloud, STOPWORDS

# Joining all the words in the description column into one string.
text = " ".join(cat for cat in df_artistas_decada["genres"])

word_cloud = WordCloud(
    width=7000,
    height=2000,
    background_color="black",
    colormap="OrRd",
    collocations=True,
    stopwords=STOPWORDS,
    ).generate(text)

# Showing the word cloud.
plt.imshow(word_cloud)


# Turning off the axis.
plt.axis("off")

# Showing the word cloud.
plt.show()
```
